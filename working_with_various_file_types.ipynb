```python
# Reading the json file
countries = []
with open("country_data.json") as file:
    countries = json.load(file)
```

```python
countries
```

```python
len(countries)
```

```python
for country in countries:
    if country.get('name') == "Albania":
        print(country)
```

```python
# Languages of each country
all_languages = []
for country in countries:
    all_languages.append(country.get('languages'))
    
all_languages
```

```python
# {"English" : 80,
# "Spanish": 78}
# flat_list = [item for sublist in l for item in sublist]
all_languages = [item for sublist in all_languages for item in sublist]

all_languages
```

```python
import collections


language_counters = collections.Counter(all_languages)
language_counters
```

```python
max(language_counters, key=language_counters.get)
```

```python
# Find the 10 most spoken languages in the world

sorted(language_counters, key=language_counters.get, reverse=True)[:10]
```

```python
from collections import Counter

c = Counter(language_counters)

most_common = c.most_common(10)
print(most_common)
```
